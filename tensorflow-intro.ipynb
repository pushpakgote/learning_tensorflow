```python
# %tensorflow_verion 2.x   ---for google colab
import tensorflow as tf
print(tf. __version__)
print(tf.version)
```

```python
string=tf.Variable("This is a string",tf.string)
num=tf.Variable(123,tf.int16)
floating=tf.Variable(2.3,tf.float16)
print(string,'\n',num,'\n',floating)
```

```python
# rank is dimension
rank1_var=tf.Variable(['hi','by','hello'],tf.string)
rank2_var=tf.Variable([['yo','one','two'],['hi','by','hello']],tf.string)
print(tf.rank(rank1_var),'\n',tf.rank(rank2_var))
```

```python
# shape : tells how many items are there in each dimension
rank2_var.shape
```

```python
#reshape
tensor1=tf.ones([1,2,3])
print('tensor1 : ',tensor1)

tensor2=tf.reshape(tensor1,[2,3,1])
print('tensor2 : ',tensor2)

tensor3=tf.reshape(tensor2,[3,-1])
print('tensor3 : ',tensor3)

tensor4=tf.reshape(tensor3,6) #Flatten
print('tensor4 : ',tensor4)
```
